## Demonstration of Underfitting and Overfitting

```python
# Import the NumPy library for numerical computations
import numpy as np

# Import Matplotlib for data visualization
import matplotlib.pyplot as plt

# Import PolynomialFeatures to create polynomial features from data
from sklearn.preprocessing import PolynomialFeatures

# Import LinearRegression for fitting the polynomial regression model
from sklearn.linear_model import LinearRegression

# Import mean_squared_error to calculate error metrics (not used in this script but often useful for evaluation)
from sklearn.metrics import mean_squared_error

# Generate synthetic data
np.random.seed(42)  # Set random seed for reproducibility
x = np.linspace(0, 10, 100).reshape(-1, 1)  # Generate 100 evenly spaced values between 0 and 10 and reshape to column vector
y = np.sin(x) + 0.1 * np.random.normal(size=x.shape)  # Add noise to sine wave to create target values

# Fit models with varying degrees of complexity
degrees = [1, 4, 15]  # List of polynomial degrees to fit models
plt.figure(figsize=(15, 5))  # Create a figure with specific size for plotting

# Loop through each degree to fit and plot polynomial models
for i, degree in enumerate(degrees):
    poly = PolynomialFeatures(degree)  # Create polynomial features of the specified degree
    x_poly = poly.fit_transform(x)  # Transform input data to polynomial features
    model = LinearRegression().fit(x_poly, y)  # Fit linear regression model to the polynomial features
    y_pred = model.predict(x_poly)  # Predict target values using the fitted model

    # Plot the data and model predictions
    plt.subplot(1, 3, i + 1)  # Create a subplot for each degree
    plt.scatter(x, y, label="Data")  # Scatter plot of the original data
    plt.plot(x, y_pred, color="red", label=f"Degree {degree}")  # Line plot of model predictions
    plt.title(f"Polynomial Degree {degree}")  # Set the title of the subplot
    plt.legend()  # Add a legend to differentiate data and predictions

plt.show()  # Display all subplots
```
